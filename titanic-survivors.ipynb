## 2.1 Pclass

* Pclass 는 ordinal, 서수형 데이터
    카테고리 이면서, 순서가 있는 데이터 타입
* Pclass 에 따른 생존율 차이
    엑셀의 피봇 차트와 유사한 작업
    pandas dataframe 에서는 grouby 를 사용하면 쉽게 할 수 있음.
    pivot 이라는 메서드도 있음.
* 아래와 같이 count()를 하면, 각 class 에 몇 명이 있는지 확인할 수 있으며,
  sum()을 하면, 216명 중 생존한(survived=1) 사람의 총합을 주게 됨.

```python
# Pclass (등석)과 Survived(DoA)을 대상으로, 등석별 카운트
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).count()
```

```python
# Pclass (등석)과 Survived(DoA)을 대상으로, 등석별 생존자 수 (사망자는 0이라서 sum()에 추가 안됨)
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).sum()
```

```python
# pandas 의 crosstab
pd.crosstab(df_train['Pclass'], df_train['Survived'], margins=True).style.background_gradient(cmap='summer_r')
```

### grouped 객체에 mean() 을 하게 되면, 각 클래스별 생존률을 얻을 수 있습니다.
### class 1 이면 아래와 같습니다.

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).mean().sort_values(by='Survived', ascending=False).plot.bar()
```

1등석, 2등석, 3등석 순으로 생존율이 높다.

```python
# seaborn 의 countplot

y_position = 1.02

f, ax = plt.subplots(1, 2, figsize=(18, 8))

df_train['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'], ax=ax[0])
ax[0].set_title('Number of Passengers By Pclass', y=y_position)
ax[0].set_ylabel('Count')

sns.countplot('Pclass', hue='Survived', data=df_train, ax=ax[1])
ax[1].set_title('Pclass: Dead or Alive', y=y_position)

plt.show()
```

* Pclass (등석)가 높을수록 생존율이 높음 (Pclass 1,2,3순으로 63%, 48%, 25% 임)
* 생존에 Pclass가 큰 영향을 미친다고 생각할 수 있으며,
   나중에 모델을 세울 때, 이 feature를 사용해야 한다고 판단할 수 있음.

## 2.3 Both Sex and Pclass

* Pclass 와 Sex 에 따라, 생존이 어떻게 달라지는지 확인한다.

* 여성이 남성 보다 생존율이 월등히 높음.

* 2등석, 3등석 남성 사망율 매우 높음.

* 성별에 관계 없이, Pclass 가 높을수록 생존율이 높지만,
        1등석 남성의 생존율이 전체 생존율 대비 크게 높다고 볼 수는 없음.
        2등석과 3등석 남성에 비하여 상대적으로 높음.

```python
# 위의 그래프는 hue 대신 column 으로 하면 아래와 같아진다.
sns.factorplot(x='Sex', y='Survived', col='Pclass', data=df_train, satureation=.5,size=9, aspect=1)
```

* 1,2등석 여성 생존율 매우 높음
* 2,3등석 남성 사망율 매우 높음
* 1등석 남성의 생존율은, 2,3등석 남성에 비하여 상대적으로 높음.

* 3등석 여성의 생존율은, 1등석 남성에 비하여 약간 높은 수준임.

## 2.4 Age

```python
print('제일 나이 많은 탑승객 : {:.1f} Years'.format(df_train['Age'].max()))
print('제일 어린 탑승객 : {:.1f} Years'.format(df_train['Age'].min()))
print('탑승객 평균 나이 : {:.1f} Years'.format(df_train['Age'].mean()))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 5))
sns.kdeplot(df_train[df_train['Survived'] == 1]['Age'], ax=ax)
sns.kdeplot(df_train[df_train['Survived'] == 0]['Age'], ax=ax)
plt.legend(['Survived == 1', 'Survived == 0'])
plt.show()
```

* 10대 중후반 기준으로,
    어린 아이 생존율이 사망율 보다 높음
    
* 10대 중후반 기준으로,
    20대 초반까지 사망율이 생존율 보다 높음.
    
* 이후 연령대는 생존율과 사망율이 비슷함.

```python
# Age distribution withing classes
plt.figure(figsize=(8, 6))
df_train['Age'][df_train['Pclass'] == 1].plot(kind='kde')
df_train['Age'][df_train['Pclass'] == 2].plot(kind='kde')
df_train['Age'][df_train['Pclass'] == 3].plot(kind='kde')

plt.xlabel('Age')
plt.title('Age Distribution within classes')
plt.legend(['1st Class', '2nd Class', '3rd Class'])
```

Pclass 가 높을수록, 중년 이후의 생존율이 높음을 알 수 있다.

Pclass 가 낮아도, 젊으면 생존율이 가장 높음을 알 수 있다.

```python
cummulate_survival_ratio = []
for i in range(1, 80):
    cummulate_survival_ratio.append(df_train[df_train['Age'] < i]['Survived'].sum() / len(df_train[df_train['Age'] < i]['Survived']))
    
plt.figure(figsize=(7, 7))
plt.plot(cummulate_survival_ratio)
plt.title('Survival rate change depending on range of Age', y=1.02)
plt.ylabel('Survival rate')
plt.xlabel('Range of Age(0~x)')
plt.show()
```

* 나이가 어릴수록 생존율이 월등히 높아짐을 알 수 있다.
* 연령이 중요한 feature 로 쓰일 수 있음.

   몇 개 없는 outlier에 대해 너무 민감하게 반응하다면, 실제 예측시에 좋지 않은 결과를 낳을 수 있음.
* outlier의 영향을 줄이기 위해, Fare 에 log를 취한다.
* pandas의 유용한 기능 사용
   dataFrame의 특정 columns에 공통된 작업(함수)를 적용하고 싶다며,
    아래의 map 또는 apply를 사용하면 쉽게 적용 가능.
* 지금 원하는 것은, Fare columns의 데이터 모두에 log값을 취하는 것인데,
  lambda 함수를 이용해 간단한 로그를 적용하는 함수를, map에 인수로 넣어주면,
  Fare columns 데이터에 그대로 적용됨.

```python
# 아래 줄은 뒤늦게 발견하였습니다. 13번째 강의에 언급되니, 일단 따라치시고 넘어가면 됩니다.
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean() # testset 에 있는 nan value 를 평균값으로 치환합니다.

df_train['Fare'] = df_train['Fare'].map(lambda i: np.log(i) if i > 0 else 0)
df_test['Fare'] = df_test['Fare'].map(lambda i: np.log(i) if i > 0 else 0)
```
